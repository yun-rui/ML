```python
import mglearn
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c=Y_train, figsize=(15,15), marker = 'o', hist_kwds={'bins':20}, s=20, alpha=.8, cmap=mglearn.cm3)
```

```python
#KNN 사용
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,Y_train)
```

```python
#레이블을 모르는 새 데이터 예측
X_new = np.array([[5,2.9,1,0.2]])
print(X_new.shape)
```

```python
prediction = knn.predict(X_new)
print("predicton result : ",prediction)
print("target name of prediction result : ",iris_dataset['target_names'][prediction])                    
```
